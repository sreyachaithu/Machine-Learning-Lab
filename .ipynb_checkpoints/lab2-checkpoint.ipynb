```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Load the dataset
file_path = 'Car_sales.csv'  # Adjust path if needed
df = pd.read_csv(file_path)

# List of numerical columns to scale
numeric_columns = [
    'Sales in thousands', '4-year resale value', 'Price in thousands',
    'Engine size', 'Horsepower', 'Wheelbase', 'Width',
    'Length', 'Curb weight', 'Fuel capacity', 'Fuel efficiency'
]

# Convert columns to numeric (coerce errors to NaN for invalid entries)
for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Drop rows with missing values in numeric columns
df_numeric = df[numeric_columns].dropna()

# Apply Standardization (Z-score)
standard_scaler = StandardScaler()
df_standard_scaled = pd.DataFrame(standard_scaler.fit_transform(df_numeric), columns=numeric_columns)

# Apply Min-Max Normalization
minmax_scaler = MinMaxScaler()
df_minmax_scaled = pd.DataFrame(minmax_scaler.fit_transform(df_numeric), columns=numeric_columns)

# Save scaled data to CSV files (optional)
df_standard_scaled.to_csv('Car_sales_standardized.csv', index=False)
df_minmax_scaled.to_csv('Car_sales_minmax_scaled.csv', index=False)

# Display first few rows
print("Standardized Data:\n", df_standard_scaled.head())
print("\nMin-Max Normalized Data:\n", df_minmax_scaled.head())
```

```python
#27Apply feature scaling on numerical variables
num=df.select_dtypes(include="number").columns
for i in num:
    std=df[i].std()
    mean=df[i].mean()
    df[i]=(df[i]-mean)/std
    print(df[i].head())   
for j in num:
    min=df[j].min()
    max=df[j].max()
    df[j]=(df[j]-max)/(max-min)
    print(df[j].head())   
```
