```python
#11. Sort the data w.r.t price_find the details of the most and the least expensive cars
sorted_order=df.sort_values(by=['Price in thousands'],ascending=True)
print("least expensive car is:\n",sorted_order.iloc[0].values)
print("most expensive car is:\n",sorted_order.iloc[-1].values)
```

```python
#12. Write a function to find min and max values of any column
#handling missing values
df['Horsepower']=pd.to_numeric(df['Horsepower'],errors='coerce')
df=df.dropna(subset= ['Horsepower'])
def min_max(df,column):
    min_value=df[column].min()
    max_value=df[column].max()
    min_row=df[ df[column] == min_value]
    max_row=df[ df[column] == max_value]
    return min_value,max_value,min_row,max_row
min_value,max_value,min_row,max_row= min_max(df,'Horsepower')
print(f"min value is:{min_value} \n",min_row.values)
print(f"max value is:{max_value} \n",max_row.values)
```

```python
#13. Call the above function to find min max of horse power, length, fuel efficiency
#handling missing values
columns=['Horsepower','Length','Fuel efficiency']
#replacing non-numeric values with null
df[columns]=df[columns].apply(pd.to_numeric,errors='coerce')
#removing rows with horsepower column having null
df=df.dropna(subset=columns)
def min_max(df,column):
    min_value=df[column].min()
    max_value=df[column].max()
    return min_value,max_value
for i in columns:
  min_value,max_value= min_max(df,i)
  print(f"minimum of {i} is:", {min_value})
  print(f"maximum of {i} is:", {max_value})
```

```python
#14. Plot histogram of continuous numerical variable :price,sales,hoursepower,fuelefficiency
import matplotlib.pyplot as plt
import seaborn as sns
columns=['Price in thousands','Sales in thousands','Horsepower','Fuel efficiency']
#handling missing values
df[columns]=df[columns].apply(pd.to_numeric,errors='coerce')
df=df.dropna(subset = columns)

fig,axes=plt.subplots(2,2 , figsize=(12,10))#2*2 grid bcz we have 4 columns and figsize for width,height of total plot
for i, column in enumerate(columns): #enumerate will give the index and column name
    ax=axes[i//2 , i%2]  # to place each plot at correct spot ex:i=0 (0,0),i=1 (0,1),i=2 (1,0)
    sns.histplot(df[column],ax=ax)
    ax.set_xlabel(column)
    ax.set_ylabel("frequency")
    ax.set_title(f"histogram for {column}")

plt.tight_layout()
plt.show()
```

```python
#15. Probability density distribution of continuous numerical variable-length
df['Length']=pd.to_numeric(df['Length'],errors='coerce')
df=df.dropna(subset=['Length'])
plt.figure(figsize=(8,5))
plt.hist(df['Length'],bins=20,density=True,alpha=0.6,color="green")
sns.kdeplot(df['Length'],fill=True ,color="blue")
plt.xlabel("Length")
plt.ylabel("density")
plt.title("Probability Density Function of Car Length")
plt.show()
```

```python
#26.Split the data set into train and test set 70% train set 10% test set
from sklearn.model_selection import train_test_split
df=pd.read_csv('Car_sales.csv')
y=df['Sales in thousands']
x=df.drop(columns=['Sales in thousands'])
x_train,x_temp,y_train,y_temp=train_test_split(x,y,test_size=0.30,random_state=42)
x_test,x_val,y_train,y_val = train_test_split(x_temp,y_temp,test_size=2/3,random_state=42)                                              
print("train data set shape is ",x_train.shape,y_train.shape)
print("test data set shape is ",x_test.shape,y_test.shape)
print("validation data set shape is ",x_val.shape,y_val.shape)
```

```python
#27. Apply feature scaling on numerical variables
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
df = pd.read_csv("Car_sales.csv")
normalizer=MinMaxScaler()
numeric=df.select_dtypes(include=['number','float64']).columns
df[numeric]=normalizer.fit_transform(df[numeric])
df
# Save the transformed dataset
df.to_csv("Car_sales_scaled.csv", index=False)

print("Feature scaling applied and saved as 'Car_sales_scaled.csv'")
```
