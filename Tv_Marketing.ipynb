```python
# If x_train and y_train are pandas Series (e.g., from a DataFrame column)

# Convert to lists to avoid .iloc or label-based indexing issues
x_train = x_train.tolist()
y_train = y_train.tolist()

# Calculate means
x_mean = sum(x_train) / len(x_train)
y_mean = sum(y_train) / len(y_train)

# Initialize numerator and denominator
num = 0
den = 0

# Loop for least squares calculation
for i in range(len(x_train)):
    num += (x_train[i] - x_mean) * (y_train[i] - y_mean)
    den += (x_train[i] - x_mean) ** 2

# Calculate slope and intercept
slope = num / den
intercept = y_mean - slope * x_mean

# Print result
print(f"Final output: y = {slope:.4f}x + {intercept:.4f}")
```
