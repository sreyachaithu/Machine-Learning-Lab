```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Load the data
df = pd.read_csv("Car_sales.csv")

# Columns to scale
columns = ["Price in thousands", "Sales in thousands", "Horsepower", "Fuel efficiency"]

# Convert to numeric and handle errors by coercing non-numeric values to NaN
df[columns] = df[columns].apply(pd.to_numeric, errors='coerce')

# Drop rows with missing values
df_cleaned = df.dropna(subset=columns)

# Apply Min-Max Scaling (Normalization)
def min_max_scale(column):
    X_min = column.min()
    X_max = column.max()
    return (column - X_min) / (X_max - X_min)

# Create a copy for normalized data
df_normalized = df_cleaned.copy()
df_normalized[columns] = df_cleaned[columns].apply(min_max_scale)

# Display normalized data
print("Normalized Data:")
print(df_normalized[columns].head())

# Apply Standardization (Z-score Normalization)
def standardize(column):
    mean = column.mean()
    std_dev = column.std()
    return (column - mean) / std_dev

# Create a copy for standardized data
df_standardized = df_cleaned.copy()
df_standardized[columns] = df_cleaned[columns].apply(standardize)

# Display standardized data
print("Standardized Data:")
print(df_standardized[columns].head())
```
