```python
#Histogram of Age of the given data set(sample)
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
df=pd.read_csv('train_titanic.csv')
df
sns.histplot(data=df,x='Age')
plt.show()
   
```

## Handling Missing Values

```python
#Replace the msiing values in Embarked column by 'S'

df['Embarked'].fillna('S', inplace=True)
print(df)
```

```python
#Impute the missing values in Fare column with median value
df['Fare'].fillna(df['Fare'].median(), inplace=True)
print(df)
```

```python
#Impute the Age column. It has too many missing values but it may be important. 
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data
import numpy as np

np.random.seed(42)  # Set seed for reproducibility

mean = df['Age'].mean()
std = df['Age'].std()
n_missing = df['Age'].isna().sum()

rand_vals = np.random.normal(mean, std, n_missing)
df.loc[df['Age'].isna(), 'Age'] = np.round(rand_vals)
print(df)
```

## Feature Engineering - Creating new features

```python
# Create and add a new feature FamilySize as a combination of SibSp and Parch
#FamilySize = Parch + SibSp + 1
df['FamilySize']=df['Parch']+df['SibSp']+1
df
```

```python
# Create new feature IsAlone from FamilySize
#If FamilySize == 1 then IsAlone = True else IsAlone = False
df['IsAlone']=df['FamilySize'].apply(lambda x: 'True' if x ==1 else 'False')
df
```

## Create a new attribute Has_cabin using the attribue Cabin. 
#if Cabin == NaN the Has_cabin = False else Has_cabin = True


```python

#Drop the columns which may not be useful in prediction of survival
#Drop PassengerId, Name, Ticket, and Cabin
s=df.drop(columns=['PassengerId', 'Name', 'Ticket','Cabin'])
s
```

## Model Building

```python
#Prepare X and y
x= df.drop(columns=['Survived'])
x
```

```python
y=df['Survived']
y
```

```python
# Label encoding of categorical features

from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
s['IsAlone']=encoder.fit_transform(s['IsAlone'])
s['Sex']=encoder.fit_transform(s['Sex'])
t=pd.get_dummies(s, drop_first=True).replace({True:1,False:0})
t
```

```python
# concat df_categorical with original df
```

```python
#Do a pairplot of all features
import seaborn as sns
sns.pairplot(t)
#plt.show(t)
```

```python
# Splitting the data into train and validation set - 90% train 10% validation
#X_train, X_test, y_train, y_validation
from sklearn.model_selection import train_test_split
X=t.drop(['Survived'], axis=1)    
y =t['Survived']
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=10)
                                                  
```

```python
# Create the model
# Fitting the decision tree with default hyperparameters
# Importing decision tree classifier from sklearn library
from sklearn.tree import DecisionTreeClassifier,plot_tree
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
#Print the tree

# Importing required packages for visualization
plot_tree(model, feature_names=X.columns, class_names=["0", "1"], filled=True)
plt.show()
```

```python
#max_depth=3
plot_tree(model, feature_names=X.columns, class_names=["0", "1"], filled=True, max_depth=3)
plt.show()
```
